```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = np.array([4, 5, 6])
x0 = 0
x_star = b # minimum since convex
def f(x):
    return np.linalg.norm(x-b)

def g(x):
    return np.linalg.norm(x-b)**2

def grad(x):
    return (x-b)/np.linalg.norm(x-b)

def grad2(x):
    return 2*(x-b)

def error(x):
    return np.linalg.norm(x - x_star) / np.linalg.norm(x_star)
```

### Task 2, constant step size

```python
err_consStep1 = []
err_consStep2 = []
num_iterations = 100
alpha_const1 = 1
alpha_const2 = 0.1
x=x0
y=x0
for k in range(num_iterations):
    x = x - alpha_const1 * grad(x)
    y = y - alpha_const2 * grad2(y)
    err_consStep1.append(error(x))
    err_consStep2.append(error(y))

plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations), err_consStep1)
plt.plot(range(num_iterations), err_consStep2)

# Find the corresponding x value for y = 0.01
corresponding_x = next(x for x, val in enumerate(err_consStep2) if val < 0.01)
print(f"For squared-nrom it reduces to 1% after {corresponding_x} many iterations")
```

### Task 3, Exponential decreasing step size

```python
err_expStep = []
err_expStep2 = []
num_iterations = 100
alpha_exp = 5/6
alpha_exp2 = 1/6
x=x0
y=x0
for k in range(num_iterations):
    x = x - (alpha_exp**k) * grad(x)
    err_expStep.append(error(x))
    y = y - (alpha_exp2**k) * grad2(y)
    err_expStep2.append(error(y))

plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations), err_expStep)
plt.plot(range(num_iterations), err_expStep2)
```

### Task 4, linearly decreasing step size

```python
err_linStep = []
err_linStep2 = []
num_iterations = 4000
alpha_lin = 1
alpha_lin2 = 1/4
x=x0
y=x0
for k in range(num_iterations):
    x = x - alpha_lin/(k+1) * grad(x)
    err_linStep.append(error(x))
    y = y - alpha_lin2/(k+1) * grad2(y)
    err_linStep2.append(error(y))

plt.figure(figsize=(10, 6))
plt.plot(range(num_iterations), err_linStep)
plt.plot(range(num_iterations), err_linStep2)

# Find the corresponding x value for y = 0.01
corresponding_x = next(x for x, val in enumerate(err_linStep) if val < 0.01)
print(f"For norm it reduces to 1% after {corresponding_x} many iterations")

# Find the corresponding x value for y = 0.01
corresponding_y = next(y for y, val in enumerate(err_linStep2) if val < 0.01)
print(f"For squared-norm it reduces to 1% after {corresponding_y} many iterations")

plt.show()
```
