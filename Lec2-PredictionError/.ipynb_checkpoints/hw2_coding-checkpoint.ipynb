```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fitPolyForLinear(degree, numPoints):
    ## Here the underlying model is linear
    z = np.random.randn(numPoints) # define it as a matrix of 1000x1 standard gaussian
    alfa = 2*np.random.rand(numPoints) - 1
    y = alfa + 1 + z
    theta = np.poly1d(np.polyfit(alfa, y, degree))
    errorLoc = ((theta(alfa) - y)**2).mean()
    return errorLoc
```

```python
def fitPolyForExp(degree, numPoints):
    ## Here the underlying model is exponential
    z = np.random.randn(numPoints) # define it as a matrix of 1000x1 standard gaussian
    alfa = 7*np.random.rand(numPoints) - 4
    y = np.exp(alfa) + z
    theta = np.poly1d(np.polyfit(alfa, y, degree))
    errorLoc = ((theta(alfa) - y)**2).mean()
    return errorLoc
```

# PROBLEM 5, WHERE WE FIT A POLYNOMIAL OF DEGREE D TO A TRUE MODEL OF LINEAR FUNCTION

```python
n = [10, 20, 50, 100, 200]
D = [2, 4, 6, 8, 10]
monte = 100
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 2)
for mm in D:
    err_list = []
    for k in n:
        summat = 0
        for i in range(monte):
            summat += fitPolyForLinear(mm, k)
        err_list.append(summat/monte)
    plt.plot(n, err_list, label = f"D = {mm}")
    
# Add labels and title
plt.xlabel('n')
plt.ylabel('Prediction Error')
plt.title('Err vs. n')

# Add a legend
plt.legend()

plt.subplot(1, 2, 1)
for k in n:
    err_list = []
    for mm in D:
        summat = 0
        for i in range(monte):
            summat += fitPolyForLinear(mm, k)
        err_list.append(summat/monte)
    plt.plot(D, err_list, label = f" n = {k}")
    
# Add labels and title
plt.xlabel('D')
plt.ylabel('Prediction Error')
plt.title('Error vs. D')

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

# PROBLEM 6: WHERE WE FIT A POLYNOMIAL OF DEGREE D TO A TRUE MODEL OF EXPONENTIAL FUNCTION

```python
n = [10, 20, 50, 120]
D = [2, 4, 6, 8]
monte = 10
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
for mm in D:
    err_list = []
    for k in n:
        summat = 0
        for i in range(monte):
            summat += fitPolyForExp(mm, k)
        err_list.append(summat/monte)
    plt.semilogy(n, err_list, label = f"D = {mm}")
    
# Add labels and title
plt.xlabel('n')
plt.ylabel('Prediction Error')
plt.title('Err vs. n')

# Add a legend
plt.legend()

plt.subplot(1, 2, 2)
for k in n:
    err_list = []
    for mm in D:
        summat = 0
        for i in range(monte):
            summat += fitPolyForExp(mm, k)
        err_list.append(summat/monte)
    plt.semilogy(D, err_list, label = f" n = {k}")
    
# Add labels and title
plt.xlabel('D')
plt.ylabel('Prediction Error')
plt.title('Error vs. D')

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

# PROBLEM 7, EXPLANATION OF THE PLOTS:
* In linear regression, when we fit a polynomial, polynomial functions can represent linear models, therefore we don't have a bias error. Thus as we solved in the question 4, the prediction error is only the part coming from the noise vector. The prediction error scales linearly with D and this can be seen on the plot of part 5, as D increases, the prediction error also increases for the same number of observations.

* However, in problem 6, polynomial fit cannot represent the exponential function; thus, we have an irreducible error (bias term), therefore prediction error decreases with increasing D value, in contrast to problem 5. However, when we increase D, overfitting to noise starts therefore, there is a trade-off between bias and variance terms.

* The variance (overfitting) will decrease with increasing n, therefore at some point bias will be the only error term.
