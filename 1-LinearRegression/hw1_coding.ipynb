### Problem 4 (Coding and Least Squares)

**Problem 4.1 and 4.2**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1000
```

```python
x = np.random.rand(n)
eps = 0.25*np.random.randn(n)
y = x + eps

plt.scatter(x, y, s=10, label="Scatter Plot")
plt.xlabel('Data')
plt.ylabel('Result')
plt.title('Scatter Plot')

a_opt = (x @ y) / (x @ x)

plt.plot(x, a_opt*x, color='red', label='g(x)=ax')
plt.legend()
plt.show()
```

**Problem 4.3 and 4.4**

```python
x = np.random.rand(n)
z = np.zeros((1000,5))

for i in range(0,n):
    z[i] = [1, x[i], x[i]**2 , x[i]**3, x[i]**4]

y = np.array(30 * (x - 0.25)**2 * (x - 0.75)**2 + 0.01 * np.random.randn(n))
axis = np.linspace(0, 1, 100).reshape(-1, 1) ** np.arange(5)

plt.scatter(x, y, s=10, label="Scatter Plot")
plt.xlabel('Data')
plt.ylabel('Result')
plt.title('Scatter Plot')

a_LS = np.linalg.pinv(z)@y
plt.plot(axis[:,1], axis@a_LS, color='red', label='g(x)=ax')
plt.legend()
plt.show()
```
